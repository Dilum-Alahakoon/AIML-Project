```python
file_path= 'https://raw.githubusercontent.com/Dilum-Alahakoon/AIML-Project/refs/heads/main/data/raw/IT24103159.csv'
Data_df= pd.read_csv(file_path)
Data_df.head()
```

Z-Secore Scaling For Numerical Columns

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Columns to scale (continuous numeric ones)
cols_to_scale = ["Age", "Hours_Per_Week", "Average_Salary"]

# Initialize scaler
scaler = StandardScaler()

# Fit and transform the selected columns
Data_df[cols_to_scale] = scaler.fit_transform(Data_df[cols_to_scale])

print(Data_df.head())
```
